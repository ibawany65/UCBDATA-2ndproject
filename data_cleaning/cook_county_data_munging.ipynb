```python
import pandas as pd
```

The "illinois_data_df.csv" file is too large to push to the github repo, so it has been removed from the folder. This is how the cleaning was done though.

```python
file1 = "illinois_data_df.csv"
file2 = "zip_code_database.csv"

ill_df = pd.read_csv(file1, dtype=str)
zip_df = pd.read_csv(file2, dtype=str)
```

```python
ill_df = ill_df.replace(to_replace='Zip Code: ', value='', regex=True)
```

```python
ill_df
```

```python
main_df = ill_df.set_index('Region').join(zip_df.set_index('zip'), how='left')
```

```python
main_df['county'].unique()
```

```python
cc_df = main_df.loc[main_df['county']=='Cook County']
```

```python
cc_df
```

```python
list(cc_df.columns)
```

```python
cc_df = cc_df.drop(columns=['Unnamed: 0', 'Worksheet Filter', 'Measure Display', 'Number of Records', 'City', 'Seasonally Adjusted',
                   'last_updated', 'months_of_supply', 'months_of_supply_mom', 'months_of_supply_yoy', 'Period Duration',
                   'Period End', 'Region Type', 'region_type_id', 'Table Id', 'type', 'decommissioned', 'acceptable_cities',
                   'unacceptable_cities', 'state', 'timezone', 'area_codes', 'world_region', 'country', 'latitude', 'longitude',
                           'Price Drops', 'Price Drops Mom', 'Price Drops Yoy', 'parent_metro_region_metro_code', 'parent_metro_region'])
```
